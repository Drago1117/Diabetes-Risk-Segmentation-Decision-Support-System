# Business and Domain Understanding
## Diabetes Risk Segmentation & Decision Support System

## 1. Introduction
BC Analytics partners with clinics, insurers and wellness programs aiming to improve patient outcomes and costs over the long term. One of the significant challenges to these various stakeholders within the healthcare space is the late identification of patients at risk of developing diabetes. In many cases, providers respond only after either symptoms have appeared or complications have started. Such a situation indicates the need for a system that can aid in both early detection while also providing better understanding of risk factors and interventions.


## 2. Business Problem
The business problem inherent in this situation is essentially the lack of such an effective tool to help healthcare providers identify those at risk for diabetes - early enough for preventative action. These providers may lack the insights regarding patients’ lifestyle and health factors associated with diabetes risk. The result is missed opportunities for such preventative measures or interventions during clinic visits.


## 3. Domain Understanding
Diabetes is essentially a category of health problems affecting the body’s ability to effectively manage glucose intake. Such problems develop either due to lack of insulin or effective use of insulin - whether either or both of those elements are at fault. The result can lead to various significant complications.
The major forms of diabetes exist in each of these categories. Type 1 diabetes problems are auto-immune in nature and produce little or no insulin use - it is not caused by lifestyle elements. Type 2 diabetes shows insulin resistance and is directly linked to lifestyle-related factors. Such instances may feature obesity, diet or physical inactivity and blood pressure issues. Prediabetes exists in a range of glucose levels that are above normal but not high enough for a diabetes diagnosis. This category is especially important as intervention at this stage can help prevent Type 2 diabetes. Gestational diabetes emerges during pregnancy yet can pose future diabetes risks.
An understanding of these categories is essential to the project. Not all forms of diabetes relate to lifestyle in the same way. This project focusses upon lifestyle-based insights and recommendations - so such an element must be considered and factored in to understand the problem.


## 4. Understanding the Target Variable: diabetes_stage
The primary target variable in this project is diabetes_stage. It is important to understand exactly what this feature represents before model development begins. Depending on the dataset, diabetes_stage may refer to diabetes type, disease progression stage, or severity category. This distinction has both clinical and analytical implications.


For example, if the target category of the project was, say, prediabetes and Type 2 diabetes, then an analysis based upon lifestyle factors would be very relevant to that goal. However, if the goal also included the inclusion of Type 1 diabetes, efforts would have to avoid suggesting lifestyle-related factors as the cause of that type of diabetes, as it is primarily autoimmune-related. Thus, the interpretation of the diabetes_stage values is necessary to ensure that any recommendations from the model are appropriate and correct.

The `diabetes_stage` variable for this dataset includes the following classes:

- Gestational: 267
- No Diabetes: 7,737
- Pre-Diabetes: 31,013
- Type 1: 117
- Type 2: 58,163

Furthermore, some values for the `diabetes_stage` category were blank.

These classes indicate that the target category for this dataset is mixed in relation to its categories. This variable includes outcomes related to no diabetes, prediabetes, Type 1, Type 2, and gestational diabetes. Thus, different recommendations can be made for patients with Type 1 diabetes than those with prediabetes, for instance. The model must be interpreted within the proper context of the patient’s diabetes stage to provide appropriate recommendations.

## 5. Business Objective
The business objective for this project is to provide healthcare providers with a tool that can help them to better understand the risk of their patient’s developing diabetes, and to provide recommendations regarding the factors that may help to reduce that risk. The business goals of the company BC Analytics intend to accomplish this goal, and to reduce the costs that are associated with treating diabetes patients.


## 6. Data Mining Objectives
The following are the objectives that will be pursued in the analysis of this data:
- To determine the diabetes stage of each patient using the diabetes_stage feature
- To compare three different classification models to determine which was the best model for predicting the diabetes stage of patients
- To determine the most important factors related to the diabetes stage
- To classify patients into three groups according to their likely lifestyle
- To present the information in a visually pleasing and interactive dashboard


## 7. Relevance of Lifestyle Factors
The factors relating to lifestyle are obviously central and central to both prediction and action. Important factors would include elements like diet, physical activity, body weight, sleep patterns, stress, smoking and alcohol and blood pressure. Such factors are especially relevant to recognizing risks of pre-diabetes and Type 2 diabetes. Understanding which factors pose the strongest risk contributes to providing more personalized recommendations to patients - those related to lifestyle factors that can impact health outcomes over the long term. The same factors, however, should be approached more cautiously in the context of Type 1 and Gestational diabetes risks - conditions with different causes.


## 8. Project Value
The project includes both clinical and business values. Clinically, it means detecting diabetes risks earlier and in a more personalized way. From a business perspective, this helps BC Analytics and partners to reduce long-term costs related to healthcare delivery - finding these high-risk individuals ahead of severe complications. The project includes value in presenting complex data and information in a relevant and interpretable manner for healthcare consultations.


## 9. Key Limitation
The major limitation exists in the target variable - the type of diabetes. Type 2 and pre-diabetes make up the bulk of records, while Type 1 and Gestational diabetes appear in very small numbers. This could impact model performance across these classes - something to consider in any evaluation. Yet because the diabetes_stage factor encompasses several distinct clinical issues, recommendations must be made carefully - to support clinical judgment and response, rather than supplanting it.


## 10. Conclusion
The project certainly does address a significant challenge in healthcare - risk prediction, segmentation according to lifestyle factors and providing actionable decisions. A deep understanding of the domain is especially important here - as is correctly interpreting the target variable of diabetes_stage and its several clinically different categories. Focusing on early detection of such problems and lifestyle factors, the project further supports BC Analytics’ mission and goals.
