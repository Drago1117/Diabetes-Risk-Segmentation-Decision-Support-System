```python
import pandas as pd

df = pd.read_csv("Diabetes_and_LifeStyle_Dataset_.csv")

df.head()
```

```python
# Remove missing values
df = df.dropna()
```

```python
# Separate features and target
X = df.drop("diabetes_stage", axis=1)

# Convert categorical → numeric
X = pd.get_dummies(X, drop_first=True)

# Ensure everything is numeric
X = X.astype(float)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Data was cleaned by removing missing values and converting categorical variables into numeric format for analysis.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)

df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
df['cluster'].value_counts()
```

```python
cluster_summary = df.groupby('cluster').mean(numeric_only=True)

cluster_summary
```

K-Means clustering (k=3) was applied to group patients based on lifestyle and health features.

```python
import shap
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42)
model.fit(X, df['diabetes_stage'])

explainer = shap.TreeExplainer(model)

shap_values = explainer(X)

shap.summary_plot(shap_values, X)
```

High-risk patients showed higher glucose levels, HbA1c, and lower physical activity. 
It is recommended to increase physical activity, improve diet, and monitor glucose levels regularly.
